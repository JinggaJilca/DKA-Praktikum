```python
import networkx as ns
import matplotlib.pyplot as plt
def show_graph(G, pos=None, title='', labels=None):
    if labels is None:
        labels = {node: node for node in G.nodes()}
    if pos is None:
        pos = ns.spring_layout(G)
    ns.draw(
        G,
        pos,
        with_labels=True,
        labels=labels,
        node_color='red',
        node_size=2000,
        font_color="white",
        font_weight="bold",
        width=3
    )
    edge_labels = ns.get_edge_attributes(G, 'weight')
    ns.draw_networkx_edge_labels(
        G,
        pos,
        edge_labels=edge_labels,
        font_color='blue',
        font_weight='bold',
        font_size=11
    )
    plt.margins(0.2)
    plt.title(title)
    plt.show()
```

```python
pos = {
    'A' : (0,3),
    'B' : (-10,2), 'C' : (10,2),
    'D' : (-15,1), 'E' : (-5,1),
    'F' : (5,1), 'G' : (15,1),
    'H' : (-18,0), 'I' : (-12,0),
    'J' : (-8,0), 'K' : (-2,0),
    'L' : (2,0), 'M' : (8,0),
    'N' : (12,0), 'O' : (18,0),
    }
G = ns.Graph()
edges= [
    ('A', 'B', 1), ('A', 'C', 3),
    ('B', 'D', 2), ('B', 'E', 1), 
    ('C', 'F', 1), ('C', 'G', 2), 
    ('D', 'H', 2), ('D', 'I', 3), 
    ('E', 'J', 3), ('E', 'K', 2), 
    ('F', 'L', 3), ('F', 'M', 2), 
    ('G', 'N', 2), ('G', 'O', 3), 
]
G.add_weighted_edges_from(edges)
```

```python
print("Elemen pada graf")
print("Daftar Node pada graf:", G.nodes())
print("Daftar Edge pada graf:", G.edges())
print("Jumlah Node pada graf:", G.number_of_nodes())
print("Jumlah Edge pada graf:", G.number_of_edges())
```

```python
show_graph(G, pos=pos)
```
