### Fill Between

```python
# Data 
x = [1, 2, 3, 4] 
y1 = [9, 20, 27, 35] 
y_main = [7, 18, 22, 29] # Data garis utama 
y2 = [6, 15, 18, 23] 
 
# Membuat figure 
fig = plt.plot(figsize=(10,6)) 
# Membuat fill_between plot 
plt.fill_between(x, y1, y2, color='skyblue', alpha=0.5, label='Fill Plot')  
# Membuat plot garis utama 
plt.plot(x, y_main, color='blue', linewidth=2, label='Garis Utama') 
# Labeling plot 
plt.title("Fill Between") 
plt.xlabel("X") 
plt.ylabel("Y") 
# Menampilkan figure 
plt.legend(loc='lower right') 
plt.show()
```

## Stack Plot (`plt.stackplot()`)

```python
data = [
    ("Budi", 81.24, 73.66, 95.89, 2023),
    ("Siti", 98.52, 84.86, 88.7, 2024),
    ("Asep", 91.96, 71.03, 79.93, 2022),
    ("Dewi", 87.96, 97.28, 71.91, 2024),
    ("Andi", 74.68, 77.76, 79.33, 2023),
    ("Lestari", 74.68, 89.88, 79.76, 2022),
    ("Joko", 71.74, 79.35, 91.89, 2022),
    ("Rina", 95.99, 85.6, 89.13, 2022),
    ("Fajar", 88.03, 86.4, 96.62, 2024),
    ("Indah", 91.24, 75.55, 84.17, 2023),
    ("Agus", 70.62, 99.09, 73.59, 2022),
    ("Fitri", 99.1, 93.25, 91.4, 2022),
    ("Rudi", 94.97, 98.18, 92.82, 2022),
    ("Maya", 76.37, 96.84, 86.84, 2024),
    ("Hadi", 75.45, 87.94, 93.13, 2024),
    ("Sri", 75.5, 97.66, 84.81, 2023),
    ("Wawan", 79.13, 72.65, 85.68, 2024),
    ("Yuni", 85.74, 75.88, 82.83, 2022),
    ("Bambang", 82.96, 71.36, 70.76, 2023),
    ("Ayu", 78.74, 79.76, 73.24, 2022),
]

#Label Tahun
lblTahun = []
for row in data:
    tahun = row[4]
    if tahun not in lblTahun:
        lblTahun.append(tahun)

lblTahun.sort()

# print(lblTahun)

#Mengelompokkan nilai per tahun dg dictionary
kelompokNilai = {}
for row in data:
    tahun = row[4]
    nilai = [row[1],row[2],row[3]]

    if tahun not in kelompokNilai:
        kelompokNilai[tahun] = nilai 
    else:
        kelompokNilai[tahun].extend(nilai)

# print(kelompokNilai)

#BOXPLOT

finalData = [kelompokNilai[thn] for thn in lblTahun ]

fig = plt.figure(figsize=(8,6))
plt.boxplot(finalData, tick_labels=lblTahun, patch_artist=True)
plt.title("Latihan Modul 5")
plt.xlabel("Tahun Angkatan")
plt.ylabel("Nilai") 
plt.show()

# SUMMARY DATA

for i in range(len(lblTahun)):
    tahun = lblTahun[i]
    nilai_tahun = finalData[i]
    
    maksimum = max(nilai_tahun)
    minimum = min(nilai_tahun)
    rata = sum(nilai_tahun) / len(nilai_tahun)

    print("Tahun : ", tahun)
    print("Nilai Maksimum : ", maksimum)
    print("Nilai Minimum : ", minimum)
    print("Rata-Rata : ", rata)
    print()


```

```python
import numpy as np
import matplotlib.pyplot as plt

# Untuk membaca file csv
dataMentah = np.genfromtxt('soal1.csv', delimiter=',', skip_header=1, dtype ='str')

# Instruksi 1: Barplot
kolomPclass = dataMentah[:,2] #Mendapatkan isi kolom Pclass (index ke 2 pada csv)
kolomSex = dataMentah[:,5] #Mendapakan isi kolom Sex (index ke 5 pada csv)
rekapGenderPerKelas = np.zeros((3,2), dtype=int) #Membuat array 2D berukuran 2 baris dan 2 kolom dengan tipe data int
daftarKelas = ['1', '2', '3']

# Melakukan perulangan sebanyak isi dari daftarKelas
for i in range(len(daftarKelas)):
    kelas = daftarKelas[i] 
    # Klasifikasi jika kelas sama dan gender sama
    getMale = (kolomPclass == kelas) & (kolomSex == 'male')
    getMFemale = (kolomPclass == kelas) & (kolomSex == 'female')
    # Menjumlahkan banyak data yang sesuai dan ditampung pada rekapGender dengan Laki-Laki pada kolom 0, sedangkan Perempuan kolom 1
    rekapGenderPerKelas[i , 0]= np.sum(getMale)
    rekapGenderPerKelas[i , 1]= np.sum(getMFemale)

# Memisahkan jumlah masing-masing gender
jumlahMale = rekapGenderPerKelas[:,0] 
jumlahFemale = rekapGenderPerKelas[:,1]
print(rekapGenderPerKelas)
kategoriKelas = ['Kelas 1', 'Kelas 2', 'Kelas 3']
x = np.arange(len(kategoriKelas)) #Menjadikan index pada sumbu X
lebarBar = 0.3 #Ketebalan batang grafik

# Instruksi 2: Box Plot
kolomSurvived = dataMentah[:,1] #Mendapatkan isi kolom survived (index ke 1 pada csv)
kolomAge = dataMentah[:,6] #Mendapatkan isi kolom age (index ke 6 pada csv)

isianUmur = (kolomAge != '') #Mendapatkan kolom umur yang tidak kosong dan ditampung pada boolean
umurValid = kolomAge[isianUmur].astype(float) #Data diubah menjadi float
statusValid = kolomSurvived[isianUmur] #Menyaring agar data berpasangan

# Klasifikasi umur yang selamat dan tidak
umurSelamat = umurValid[statusValid == '1']
umurTidakSelamat = umurValid[statusValid == '0']

# Set atribut yang dibutuhkan boxplot
dataBoxplot = [umurSelamat, umurTidakSelamat]
lblBoxplot = ['Selamat', 'Tidak Selamat']

# Membuat Subplot
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# === KANVAS KIRI ( ax[0] ) ===
ax[0].bar(x - lebarBar/2, jumlahMale, width=lebarBar, hatch='/', color='skyblue', label='Laki-laki') 
ax[0].bar(x + lebarBar/2, jumlahFemale, width=lebarBar, hatch='\\', color='pink', label='Perempuan') 

ax[0].set_title("Jumlah Penumpang per Kelas dan Gender", fontweight='bold')
ax[0].set_xlabel("Kategori Kelas Tiket")
ax[0].set_ylabel("Jumlah Penumpang")
ax[0].set_xticks(x)
ax[0].set_xticklabels(kategoriKelas)
ax[0].legend()

# === KANVAS KANAN ( ax[1] ) ===
ax[1].boxplot(dataBoxplot, tick_labels=lblBoxplot, patch_artist=True)

ax[1].set_title("Distribusi Usia berdasarkan Status Keselamatan", fontweight='bold')
ax[1].set_xlabel("Status Keselamatan")
ax[1].set_ylabel("Usia (Tahun)")
ax[1].grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Untuk membaca file csv
dataMentah = np.genfromtxt('soal2.csv', delimiter=',', skip_header=1, dtype ='str')

# Mendapatkan isi kolom
rawKolomMat = dataMentah[:,5] 
rawKolomRead = dataMentah[:,6] 

# Membersihkan tanda kutip dan mengubah tipe data menjadi int
kolomMath = np.char.replace(rawKolomMat, '"', '').astype(int)
kolomRead = np.char.replace(rawKolomRead, '"', '').astype(int)

#=== HISTOGRAM ===
fig = plt.figure(figsize=(10,8))
plt.hist(kolomMath, bins=20, color="blue", alpha=0.7, label="Math Score") 
plt.hist(kolomRead, bins=20, color="red", alpha=0.5, label="Reading Score") 

# Labeling  
plt.title("Histogram Math Score & Reading Score", fontweight='bold') 
plt.xlabel("Nilai Ujian") 
plt.ylabel("Frekuensi (Jumlah Siswa)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()


# === BOXPLOT ===
fig = plt.figure(figsize=(10, 8))

# Menyiapkan data untuk boxplot
dataBoxplot = [kolomMath, kolomRead]
lblBoxplot = ['Math Score', 'Reading Score']

plt.boxplot(dataBoxplot, tick_labels=lblBoxplot, patch_artist=True)

# Labeling 
plt.title("Perbandingan Distribusi Math Score dan Reading Score", fontweight='bold') 
plt.ylabel("Nilai") 
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
