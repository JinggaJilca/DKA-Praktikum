## Import Library ##

```python
import matplotlib.pyplot as plt 
import networkx as nx
```

```python
G = nx.Graph()

G.add_edges_from([
    ("Jakarta", "Bandung"),
    ("Jakarta", "Surabaya"),
    ("Bandung", "Yogyakarta"),
    ("Yogyakarta", "Surabaya"),
    ("Surabaya", "Bali"),
    ("Bandung", "Bali")
])

pos = {
    "Jakarta": (0, 0),
    "Bandung": (1, 1),
    "Yogyakarta": (2, 1),
    "Surabaya": (2, 0),
    "Bali": (3, 0.5)
}

nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000)
plt.title("Graph Rute Travel")
plt.show()

#BFS
bfs_path = nx.shortest_path(G, source="Jakarta", target="Bali")
print("Jalur BFS (terpendek):", bfs_path)

#DFS
dfs_edges = list(nx.dfs_edges(G, source="Jakarta"))
print("Urutan DFS:", dfs_edges)
```

## Membuat Graf Kosong ##

```python
pos = {'A': [ 0.93888004, -0.0449161 ], 
'B': [-0.73226196, -0.53345128], 
'C': [0.29334637, 1.        ], 
'D': [-0.79991428,  0.56134392], 
'E': [ 0.29994983, -0.98297654]}

def show_graph(G, pos=None, title='') : 
  if pos == None : 
    pos = nx.spring_layout(G) 
 
  nx.draw( 
      G, 
      pos, 
      with_labels=True, 
      node_color='red', 
      node_size=2000, 
      font_color="white", 
      font_weight="bold", 
      width=5 
  ) 
 
  edge_labels = nx.get_edge_attributes(G, 'weight') 
  nx.draw_networkx_edge_labels( 
      G, 
      pos, 
      edge_labels=edge_labels, 
      font_color='blue', 
      font_weight="bold", 
      font_size=12, 
  ) 
 
  plt.margins(0.2) 
  plt.title(title) 
  plt.show() 
```

```python
G_undirected = nx.Graph()

show_graph(G_undirected, title = "Undirected Graph kosong")
```

```python
G_directed = nx.DiGraph()

show_graph(G_directed, title = "Directed Graph kosong")
```

### TUGAS MANDIRI 3

```python
import networkx as nx 
import matplotlib.pyplot as plt

#Membuat wadah kosong untuk menyimpan graph
G_undirected = nx.Graph()

# Menambahkan Node
node_data = ['A', 'B', 'C', 'D', 'E'] 
#Memasukkan ke dalam graph
G_undirected.add_nodes_from(node_data)

# Menambahkan Edge
edge_data = [('A', 'B'), ('A', 'C'), ('B', 'D'), ('C', 'D'), ('D', 'E')] 

# Memasukkan edge ke graph
G_undirected.add_edges_from(edge_data) 

# Membuat ffungsi untuk menampilkan graph
def show_graph(G, pos=None, title=""):
    # Percabangan jika posisi tidak diset diawal, maka akan otomatis dibuat layout oleh spring_layout()
    if pos == None:
        pos = nx.spring_layout(G)

    # Menggambar graph
    nx.draw( 
        G, 
        pos, 
        with_labels=True, 
        node_color='blue', 
        node_size=2000, 
        font_color="white", 
        font_weight="bold", 
        width=5 
    )
    # Mengatur margin
    plt.margins(0.2) 
    # Menampilkan title
    plt.title(title) 
    # Menampilkan graph
    plt.show() 

# Memanggil fungsi
show_graph(G_undirected, None, "Soal 1 Tugas Mandiri 3")

# Daftar Degre
print()
print("2.1 Hitung degree setiap Node")
# Melakukan perulangan sebanyak node dan mendapatkan derajat tiap node
for node, degree in G_undirected.degree():
    print(f"Node {node} memiliki derajat: {degree}")
print()
print("2.2 Tetangga node D")
# Melakukan perulangan untuk mendapatkan isi dari tetangga D
i = 0
for neighbor in G_undirected.neighbors('D'):
    print(f"D -> {neighbor}")
    i = i+1
print("Total Tetangga D: ", i)


print()
print("3. Tetangga node D")
# Mendapatkan cyle melalui fungsi bawaan
cycle = nx.find_cycle(G_undirected)
# Mencetak isi cycle
print("Graf memiliki siklus:", cycle)
# Mendapatkan basis node basis cycle
print('Daftar Cycle: ', nx.cycle_basis(G_undirected))

print("Terdapat bentuk yang berupa lingkaran yang saling terhubung dan hal ini terbukti pada node A-B-C-D")

print()
print("4. Node paling penting")
print("Menurut saya, node D karena memiliki degree paling banyak = 3 dan node ini sebagai penghubung node C")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Membuat wadah kosong untuk graph undirected
G_undirected = nx.Graph()

# Inisialisasi Weight tiap node (Hapus duplikasi C-D)
edge_weight_data = [
    ('A', 'B', 4), 
    ('A', 'C', 2), 
    ('B', 'C', 1), 
    ('B', 'D', 5), 
    ('C', 'D', 8), 
    ('C', 'E', 1), 
    ('D', 'E', 2)
]

# Menambahkan edges dengan weight
G_undirected.add_weighted_edges_from(edge_weight_data)

# Tentukan posisi node TERLEBIH DAHULU (Penting!)
pos = nx.spring_layout(G_undirected) 

# Menggambar node dan edge
nx.draw(G_undirected,
        pos, 
        with_labels=True, 
        node_color='blue', 
        node_size=2000, 
        font_color="white", 
        font_weight="bold", 
        width=3
)



# Mengambil atribut lalu set pada graph
edge_labels = nx.get_edge_attributes(G_undirected, 'weight')


# Menggambar label bobot pada edges
nx.draw_networkx_edge_labels(G_undirected, pos, edge_labels=edge_labels, font_size=10)

# Menampilkan title
plt.title("Skenario 2: Soal 2 Tugas Mandiri 3") 

# Menampilkan graph
plt.show() 


print()
print("2. Jalur dengan total weight terkecil A -> E")
# Mendapatkan daftar node yang dilalui dari A-E yang paling pendek
shortest_paths = nx.shortest_path(G_undirected, source='A', target='E', weight='weight')
# Mendapatkan total weight yang kecil dilalui dari A-E 
shortest_total_weight = nx.shortest_path_length(G_undirected, source='A', target='E', weight='weight')

print(f"Jalur A -> E : {'-> ' .join(shortest_paths)}")
print(f"Total Weight : {shortest_total_weight}")


print()
print("3. Analisis")
print("3.1 Jalur Optimal: Iya, karena memiliki weight paling rendah sehingga tidak mengeluarkan cost banyak")
print("3.2 Peran Weight: Sebagai biaya atau cost yang dikeluarkan, sedangkan tujuan jalur optimal adalah memiliki cost paling minimum sehingga dapat dikatakan efisien")

print()
print("4. Skenario weight (C,E) = 1")
print("Jawaban berubah karena secara otomatis akan memilih edges dengan bobot paling minimum dan hasil menunjukkan step yang ringkas")
```
