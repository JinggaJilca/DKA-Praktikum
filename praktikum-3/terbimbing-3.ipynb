### Latihan Soal 1

```python
import networkx as nx
import matplotlib.pyplot as plt

#Membuat wadah kosong untuk graph undirected
G_undirected = nx.Graph()

# Inisialisasi Weight tiap node
edge_weight_data = [
    ('Oradea', 'Zerind', 71), 
    ('Oradea', 'Sibiu', 151),
    ('Zerind', 'Arad', 75), 
    ('Arad', 'Sibiu', 140),
    ('Arad', 'Timisoara', 118), 
    ('Timisoara', 'Lugoj', 111),
    ('Lugoj', 'Mehadia', 70), 
    ('Mehadia', 'Drobeta', 75),
    ('Drobeta', 'Craiova', 120), 
    ('Craiova', 'Rimnicu Vilcea', 146),
    ('Craiova', 'Pitesti', 138), 
    ('Rimnicu Vilcea', 'Sibiu', 80),
    ('Rimnicu Vilcea', 'Pitesti', 97), 
    ('Sibiu', 'Fagaras', 99),
    ('Fagaras', 'Bucharest', 211), 
    ('Pitesti', 'Bucharest', 101),
    ('Bucharest', 'Giurgiu', 90), 
    ('Bucharest', 'Urziceni', 85),
    ('Urziceni', 'Vaslui', 142), 
    ('Urziceni', 'Hirsova', 98),
    ('Hirsova', 'Eforie', 86), 
    ('Vaslui', 'Iasi', 92),
    ('Iasi', 'Neamt', 87)
]

#Membuat daftar untuk agar saling terhubung
G_undirected.add_weighted_edges_from(edge_weight_data)

#Menentukan posisi sesuai gambar menggunakan koordinat
pos = {
    'Oradea': (1.5, 10), 
    'Zerind': (1, 8.5), 
    'Arad': (0.5, 7),
    'Timisoara': (0.5, 4.5), 
    'Lugoj': (1.5, 3.5), 
    'Mehadia': (1.5, 2.3),
    'Drobeta': (1.5, 1), 
    'Sibiu': (3, 6), 
    'Rimnicu Vilcea': (3.5, 5),
    'Craiova': (4, 1), 
    'Fagaras': (5, 6), 
    'Pitesti': (5.5, 4),
    'Bucharest': (7, 2.5), 
    'Giurgiu': (6.5, 1), 
    'Urziceni': (8, 3),
    'Vaslui': (9, 6), 
    'Iasi': (8.5, 8), 
    'Neamt': (7.5, 9),
    'Hirsova': (9.5, 3), 
    'Eforie': (10, 1.5)
}

#Set ukuran framenya
plt.figure(figsize=(12, 8))

#Menggambar node
nx.draw(G_undirected, pos, 
            with_labels=True, 
            node_color='lightgrey', 
            node_shape='s',
            node_size=2500,
            font_size=8,
            font_weight='bold',
            edge_color='black',
            linewidths=2)

#Data weight
edge_labels = nx.get_edge_attributes(G_undirected, 'weight')

#Tulisan bobot pada edges
nx.draw_networkx_edge_labels(G_undirected, pos, edge_labels=edge_labels, font_size=8)

plt.title("Latihan 1 Modul 3")
plt.axis('off')
plt.show()



#Membuat dictionary data rute terpendek
semua_rute = dict(nx.all_pairs_shortest_path(G_undirected))

print("=== SHORTEST PATH ===")
nomor = 1
#Perulangan untuk setiap kota asal
for kota_asal, daftar_tujuan in semua_rute.items():
    #Perulangan untuk setiap kota tujuan dari asal tersebut
    for kota_tujuan, jalur in daftar_tujuan.items():
        
        #Mencetak jika kota asal dan tujuan tidak sama
        if kota_asal != kota_tujuan:
            #Menggabungkan list dengan '->'"
            rute_teks = " -> ".join(jalur)
        
            print(f"{nomor:2}. {kota_asal:10} ke {kota_tujuan:10} : {rute_teks}")
            nomor += 1
```

### Latihan Soal 2

```python
import networkx as nx
import matplotlib.pyplot as plt

#Membuat wadah kosong untuk graph undirected
G_undirected = nx.Graph()

# Inisialisasi Weight tiap node
edge_weight_data = [
    ('Jakarta', 'Cirebon', 327), 
    ('Jakarta','Bandung', 270),
    ('Cirebon','Bandung', 120),
    ('Cirebon','Semarang', 305),
    ('Cirebon', 'Yogyakarta', 210),
    ('Bandung','Yogyakarta', 373),
    ('Semarang','Surabaya', 369),
    ('Semarang','Surakarta', 97),
    ('Semarang','Yogyakarta', 109),
    ('Yogyakarta','Surakarta', 60),
    ('Surakarta','Malang', 370),
    ('Malang','Surabaya', 94)
]

#Membuat daftar untuk agar saling terhubung
G_undirected.add_weighted_edges_from(edge_weight_data)

#Menentukan posisi sesuai gambar menggunakan koordinat
# maks y = 10

pos = {
    'Jakarta': (1, 10), 
    'Bandung': (3, 3), 
    'Cirebon': (5, 8), 
    'Yogyakarta': (8, 1), 
    'Semarang': (8.7, 7.5), 
    'Surakarta': (11, 4), 
    'Malang': (15, 1.5), 
    'Surabaya': (18, 8), 

}

#Set ukuran framenya
plt.figure(figsize=(10, 5))

#Menggambar node
nx.draw(G_undirected, pos, 
            with_labels=True, 
            node_color='lightgrey', 
            node_shape='o',
            node_size=2500,
            font_size=8,
            font_weight='bold',
            edge_color='black',
            linewidths=2)

#Data weight
edge_labels = nx.get_edge_attributes(G_undirected, 'weight')

#Tulisan bobot pada edges
nx.draw_networkx_edge_labels(G_undirected, pos, edge_labels=edge_labels, font_size=8)

plt.title("Latihan 2 Modul 4")
plt.axis('off')
plt.show()



#Membuat dictionary data rute terpendek
semua_rute = dict(nx.all_pairs_shortest_path(G_undirected))

print("=== SHORTEST PATH ===")
nomor = 1
#Perulangan untuk setiap kota asal
for kota_asal, daftar_tujuan in semua_rute.items():
    #Perulangan untuk setiap kota tujuan dari asal tersebut
    for kota_tujuan, jalur in daftar_tujuan.items():
        
        #Mencetak jika kota asal dan tujuan tidak sama
        if kota_asal != kota_tujuan:
            #Menggabungkan list dengan '->'"
            rute_teks = " -> ".join(jalur)
        
            print(f"{nomor:2}. {kota_asal:10} ke {kota_tujuan:10} : {rute_teks}")
            nomor += 1
```
