### SOAL 1

```python
import networkx as nx
import matplotlib.pyplot as plt

# Inisialisasi graf kosong 
Graph1 = nx.Graph() 

# Definisi edge sesuai dengan gambar 
edges = [ 
    ('A', 'B'), ('A', 'C'),  
    ('B', 'D'), ('B', 'E'),  
    ('C', 'F'), ('C', 'G'),  
] 
 
# Menambahkan edge pada graf 
Graph1.add_edges_from(edges) 

#Membuat posisi
pos = { 
    'A': (0, 3), 
    'B': (-10, 2), 'C': (10, 2), 
    'D': (-15, 1), 'E': (-5, 1), 'F': (5, 1), 'G': (15, 1)
} 

# Fungsi pendukung untuk mencetak graf 
def show_graph(G, pos=None, title='') : 
  # Membuat pos jika pos tidak diberikan 
  if pos is None: 
      pos = nx.spring_layout(G)  
   
  # Fungsi untuk menggambar node 
  nx.draw( 
      G,                  # Graf NetworkX 
      pos,                # Posisi node 
      with_labels=True,   # Menampilkan nama node 
      node_color='red',   # Warna node 
      node_size=2000,     # Ukuran node 
      font_color="white", # Warna font label node 
      font_weight="bold", # Ketebalan font label node 
      width=3             # Ketebalan garis edge 
  ) 
 
  # Mengambil label edge jika ada weight 
  edge_labels = nx.get_edge_attributes(G, 'weight')   
  # Fungsi untuk menggambar node 
  nx.draw_networkx_edge_labels( 
      G, 
      pos, 
      edge_labels=edge_labels, # Data weight 
      font_color='blue',       # Warna font label edge 
      font_weight="bold",      # Berat font label edge 
      font_size=12,            # Ukuran font label edge 
  ) 
 
  plt.margins(0.2)                  # Memberikan margin pada plot 
  plt.title("Soal 1 Mandiri 5")     # Menampilkan judul graf jika diberikan
  plt.show()                        # Menampilkan graf menggunakan matplotlib 

# Memanggil fungsi showGraph()
show_graph(Graph1, pos=pos) 

# BFS
print("Urutan BFS (Edges):") 
i = 1 #Sebagai index
for edge in nx.bfs_edges(Graph1, source='A'):   #Melakukan perulangan sebanyak node dengan fungsi default bfs.edges untuk menjelajahi level per level dengan sumber awal 'A'
  print(f"{i}. {edge[0]} - {edge[1]}")          #Mencetak isi edge sepasang
  i  += 1                                       #Melakukan increment nilai i

print()
print("Urutan BFS (Node):")
print("1. A")
i = 2
for nodes in nx.bfs_edges(Graph1, source='A'): 
  print(f"{i}. {nodes[1]}")                     #Untuk mendapatkan node sasaran
  i  += 1 
```

### SOAL 2

```python
import networkx as nx
import matplotlib.pyplot as plt
 
#Membuat wadah kosong untuk graph undirected
Graph2 = nx.Graph()
 
# Inisialisasi Weight tiap node
edgeWeight = [
    ('Bandung', 'Jakarta', 150),
    ('Bandung', 'Yogyakarta', 380),
    ('Jakarta', 'Semarang', 450),
    ('Yogyakarta', 'Surabaya', 330),
    ('Semarang', 'Surabaya', 350),

]
 
#Membuat daftar untuk agar saling terhubung
Graph2.add_weighted_edges_from(edgeWeight)
 
 

# Fungsi pendukung untuk mencetak graf 
def show_graph(G, pos=None) : 
  # Membuat pos jika pos tidak diberikan 
  if pos is None: 
      pos = nx.spring_layout(G)  
   
  # Fungsi untuk menggambar node 
  nx.draw( 
      G,                  # Graf NetworkX 
      pos,                # Posisi node 
      with_labels=True,   # Menampilkan nama node 
      node_color='red',   # Warna node 
      node_size=2500,     # Ukuran node 
      font_color="white", # Warna font label node 
      font_weight="bold", # Ketebalan font label node 
      font_size= 8,
      width=2           # Ketebalan garis edge 
  ) 
 
  # Mengambil label edge jika ada weight 
  edge_labels = nx.get_edge_attributes(G, 'weight')   
  
  # Fungsi untuk menggambar node 
  nx.draw_networkx_edge_labels( 
      G, 
      pos, 
      edge_labels=edge_labels,  # Data weight 
      font_color='black',       # Warna font label edge 
      font_size=8,              # Ukuran font label edge 
  ) 
  #Tulisan bobot pada edges
  nx.draw_networkx_edge_labels(Graph2, pos, edge_labels=edge_labels, font_size=8)
  
  plt.margins(0.2)  # Memberikan margin pada plot 
  plt.title("Soal 2 Mandiri 5")  # Menampilkan judul graf jika diberikan
  plt.show()        # Menampilkan graf menggunakan matplotlib 



# Memanggil fungsi showGraph()
show_graph(Graph2) 
 
# BFS  
print("Urutan BFS Transversal:") 
i = 1 
for edge in nx.bfs_edges(Graph2, source='Bandung'):    #Untuk menjelajahi node per level dengan nx.bfs_edges dengan awalan 'Bandung'
  print(f"{i}. {edge[0]} - {edge[1]}")                #Mencetak node asal dan sasarannya (tujuan)
  i  += 1 
```
